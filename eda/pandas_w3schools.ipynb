```python
import pandas as pd
```

```python
mydataset = {
  'cars': ["BMW", "Volvo", "Ford"],
  'passings': [3, 7, 2]
}

myvar = pd.DataFrame(mydataset)

print(myvar)
```

```python
a = [1, 7, 2]
myvar = pd.Series(a)
print(myvar)
```

```python
a = [11, 71, 21]
myvar = pd.Series(a, index = ["x", "y", "z"])
print(myvar)
```

```python
calories = {"day1": 420, "day2": 380, "day3": 390}
myvar = pd.Series(calories)
print(myvar)
```

```python
data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

myvar = pd.DataFrame(data, index = ["a", "b", "c"])
print(myvar)
```

```python
#refer to the row index:
print(myvar.loc['a'])
```
